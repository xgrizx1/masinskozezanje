```python
import cv2
import numpy as np
import os
from PIL import Image
from keras import models
from keras import backend as K
K.set_image_dim_ordering('tf')


print(os.listdir(os.curdir))

#Load the saved model
model = models.load_model('my_model1.h5')
video = cv2.imread('img_4.jpg', 0)

while True:
	#_, frame = video.read()
	#frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
	#Convert the captured frame into RGB
	#im = Image.fromarray(video, 'L')
	#Resizing into 128x128 because we trained the model with this image size.
	im = Image.fromarray(video)  
	im = im.resize((28,28))  
	img_array = np.array(im)
	img_array.transpose(1, 0) 

	#Our keras model used a 4D tensor, (images x height x width x channel)
	#So changing dimension 28x28x3 into 1x28x28x3 
	img_array = np.expand_dims(img_array, axis=0)
	img_array = np.expand_dims(img_array, axis=3)
    
    
	#print(img_array)
	#Calling the predict method on model to predict 'me' on the image
	prediction = int(model.predict(img_array)[0][0])
	
	#if prediction is 0, which mean I am missing on the image, then show the frame in gray color.
	print(prediction)  
	#cv2.imshow("Capturing", frame)
	key=cv2.waitKey(1)
	if key == ord('q'):
		break
video.release()
cv2.destroyAllWindows()
```
